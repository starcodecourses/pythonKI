Um das ganze zugnglicher zu machen, knnen wir den Graphen auch noch ein bisschen vergrern. Dafr benutzen wir den Befehl `plt.figure()`. Dieser erstellt ein leeres Bild. Das Ganze knnt ihr euch wie eine unbemalte Leinwand oder ein unbedrucktes Blatt Papier vorstellen. Da wir natrlich die Leinwand oder das Papier vor dem Malen oder Drucken brauchen, muss dieser Befehl auch vor allen anderen stehen. Die Gre kann dann mit dem Argument `figsize` angepasst werden. Die Gre wird dabei durch `(B,H)` dargestellt, wobei `B` die Breite und `H` die Hhe ist. Matplotlib fllt dann die gesamte Leinwand bis zum Rand aus! So sieht das dann aus:

```python
plt.figure(figsize=(14,9)) # Dieser Befehl steht immer vor allen Befehlen die die Visualisierung betreffen

plt.plot()
plt.xlabel('x-Achsen Beschriftung')
plt.ylabel('y-Achsen Beschriftung')
plt.title('berschrift')

plt.show() # Dieser Befehl steht immer nach allen anderen Befehlen die Visualisierung betreffen
```

Das einzige Problem das uns noch bleibt ist, dass zwar das Bild grer ist, aber die Achsenbeschriftung nicht. Dies lsst sich durch das Argument `fontsize` (also Schriftgre auf englisch) beheben.

## <span style="color: #9C68C8;">Streudiagramme (Scatterplots)</span> 

Ein weiterer ntzlicher Diagrammtyp ist das Streudiagramm (auch Punktwolke genannt)! Dabei werden Punkte anstelle von Linien verwendet, um Daten darzustellen. Diese Art von Diagrammen wird oft benutzt, um sogenannte 'Cluster' zu finden, also Subgruppen von Datenpunkten, die etwas gemeinsam haben. Manchmal mchte man aber auch einfach Zusammenhnge oder Tendenzen erkennen. Wir werden das gleich an einem Beispiel besser verstehen knnen!

Um so ein Diagramm zu erstellen, knnen wir den Befehl `plt.scatter()`  benutzen. Das Coole an Matplotlib ist, dass wir fast alles, was wir bisher gelernt haben, genauso anwenden knnen! 

Schauen wir uns mal ein Beispiel an: Wir plotten die Dichte in Abhngigkeit vom Durchmesser der Planeten in unserem Sonnensystem:

```python
# Reihenfolge: Merkur, Venus, Erde, Mars, Jupiter, Saturn, Uranus, Neptun
Durchmesser = [4879, 12104, 12756, 6792, 142984, 120536, 51118, 49528] # in km
Dichte      = [5429, 5243, 5514, 3934, 1326, 687, 1270, 1638]          # in kg/m^3

plt.figure(figsize=(14,9))
plt.scatter(Durchmesser, Dichte)
plt.title('Zusammenhang zwischen Dichte und Durchmesser', fontsize=30)
plt.xlabel('Durchmesser d[km]', fontsize=25)
plt.ylabel('Dichte [kg/m^3]', fontsize=25)
plt.show()
```

Zustzlich knnen wir noch die Gre der Punkte mit `s` (fr *size*) anpassen, und das aussehen der Punkte mit `marker` anpassen. Welche Marker zur verfgung stehen kann  in der Dokumenatation nachgeschlagen werden: https://matplotlib.org/stable/api/markers_api.html.

### <span style="color: #C26BCA;"> Balkendiagramme </span>

Manchmal mchten wir statt Liniendiagrammen Balkendiagramme verwenden, um Daten zu vergleichen. Diese Form der Darstellung ist besonders geeignet fr Datenstze mit wenigen Datenpunkten. Stell dir vor, du hilfst bei deinem Lieblingseisladen im Sommer aus. Um zu wissen welche Eissorten in groen Mengen produziert werden mssen, schaut ihr euch die Verkaufszahlen von diesem und letzten Jahr an:

```python
import numpy as np
import matplotlib.pyplot as plt

geschmcker = ['Vanille', 'Schokolade', 'Stracciatella', 'Erdbeere', 'Mango']
verkauft_letztes_jahr = [459, 189, 391, 303, 652]
verkauft_dieses_jahr = [679, 575, 677, 146, 801]

plt.figure(figsize=(14,9))

breite = 0.35  # Breite der Balken 
x = np.arange(len(geschmcker))  # np.arrange funktioniert wie range() nur fr Gleitkommazahlen

# Die Argumente sind bar(Position der Balken (Mitte des Balkens), Y-Wert, Breite der Balken, Label)
plt.bar(x - breite/2, verkauft_letztes_jahr, width=breite, label='Vorheriges Jahr')
plt.bar(x + breite/2, verkauft_dieses_jahr,  width=breite, label='Dieses Jahr')

plt.xlabel('Eissorten', fontsize=25)
plt.ylabel('Verkaufszahlen', fontsize=25)
plt.title('Eisverkauf im Vergleich', fontsize=30)

# Ticks bezeichnet die Markierungen (oder Teilstriche) an der jeweiligen Achse
plt.xticks(x, geschmcker, fontsize=20) # Dieser Befehl ermglicht anstelle von Zahlen, die Geschmcker darzustellen. Dabei wird an der Position in 'x' der entsprechende String in 'Geschmcker' angegeben
plt.xticks(fontsize=20) # Hier vergrern wir nur die kleinen Marker an der y-Achse

plt.legend(fontsize=20)
plt.show()
```

### <span style="color: #C26BCA;"> Histogramme </span>

Histogramme helfen uns, die Verteilung von Daten zu visualisieren. Man kann sich ein Histogramm folgendermaen vorstellen: Zunchst ermittelt man die kleinste und die grte Zahl in einem Datensatz und teilt den Bereich dazwischen in eine festgelegte Anzahl gleich groer Intervalle auf (sogenannte `bins`, engl fr *Behlter*). Jeder Datenpunkt, der sich in das entsprechende Intervall einordnet, wird in dieses Intervall einsortiert. Am Ende zhlt man die Anzahl der Elemente in jedem Intervall. Die ensprechende Hufigkeit stellen wir dann mit Balken dar. Nehmen wir zum Beispiel die Zeiten die 30 Schlerinnen im 100-Meter-Lauf gebraucht haben.
