## Herstellung von Arrays

Wie wir im vorherigen Beispiel gesehen haben, knen wir ein 1-dimensionales Array haben. Das bedeutet, dass wir nur ein Element im Array haben, das 0-D ist. 

Wir knnen aber auch multi-dimensionale Arrays haben. In vielen Wissenschaften ist es recht sinnvoll multidimensionale Arrays zu haben. 

Stellt euch vor ihr wrdet gerne unser Sonnensystem in Code reprsentieren. Ein 2D-Array kann verwendet werden, um die Positionen und Geschwindigkeiten der Planeten in unserem Sonnensystem darzustellen. Jede Zeile im Array reprsentiert einen Planeten und die Spalten reprsentieren die x-, y- und z-Koordinaten.

````python
# Definiere das 2D-Array, um das Sonnensystem darzustellen

sonnen_system = [
    # x-, y- und z-Koordinaten fr Merkur
    [0.387, 0, 0],
    # x-, y- und z-Koordinaten fr Venus
    [0.723, 0, 0],
    # x-, y- und z-Koordinaten fr die Erde
    [1, 0, 0],
    # x-, y- und z-Koordinaten fr Mars
    [1.524, 0, 0],
    # x-, y- und z-Koordinaten fr Jupiter
    [5.203, 0, 0],
    # x-, y- und z-Koordinaten fr Saturn
    [9.539, 0, 0],
    # x-, y- und z-Koordinaten fr Uranus
    [19.18, 0, 0],
    # x-, y- und z-Koordinaten fr Neptun
    [30.07, 0, 0]
]

# Gib die Position jedes Planeten im Sonnensystem aus
for i, planet in enumerate(sonnen_system):
    print(f"Planet {i+1}: x={planet[0]}, y={planet[1]}, z={planet[2]}")
````

Dieser Code definiert eine 2D-Liste **sonnen_system**, in der jede Zeile einen Planeten darstellt und die Spalten die x-, y- und z-Koordinaten dieses Planeten darstellen. Anschlieend wird eine for-Schleife verwendet, um die Position jedes Planeten im Sonnensystem auszudrucken.

## Slicing Arrays 

Zum Slicing eines Arrays verwenden wir in Python einen Bereich (englisch "slice") zwischen zwei gegebenen Indizes.

Wir bergeben den Bereich anstelle eines Indexes, und zwar wie folgt: [start:end].

Wir knnen auch das Intervall definieren, in dem wir Schritte machen mchten, wie folgt: [start:end:step].

Wenn wir keinen Startwert angeben, wird er automatisch auf 0 gesetzt.

Wenn wir keinen Endwert angeben, wird die Lnge des Arrays in dieser Dimension automatisch als Endwert betrachtet.

Wenn wir keinen Schritt angeben, wird dieser automatisch als 1 gesetzt.

Ausgabe des letzten Elements aus der 2. Dimension:
Um auf das letzte Element der 2. Dimension zuzugreifen, knnen wir die negative Indizierung verwenden:

````python
# Ausgabe des letzten Elements aus der 2. Dimension:
#Um auf das letzte Element der 2. Dimension zuzugreifen, knnen wir die negative Indizierung verwenden:

array[-1][-1]
````

````python
# Beim Slicing eines Arrays nehmen wir Elemente von einem gegebenen Index bis zu einem anderen gegebenen Index. 
# Wir verwenden dazu die "slice"-Notation mit Doppelpunkt:

array[start:end]
````

Dabei steht "start" fr den Index des ersten Elements, das wir in unseren Slice aufnehmen wollen, whrend "end" fr den Index des ersten Elements steht, das wir nicht mehr in unseren Slice aufnehmen mchten.

Es ist auch mglich, das Intervall festzulegen, in dem wir Schritte machen mchten, indem wir einen dritten Parameter "step" verwenden:

````python
array[start:end:step]
````

Wenn wir keine Start- oder Endwerte angeben, wird das Array in dieser Dimension von Anfang bis Ende durchlaufen. Wenn wir keinen Schritt angeben, wird er automatisch auf 1 gesetzt.

## Iteration von Arrays

In NumPy kann man auf verschiedene Weisen durch die Elemente eines Arrays iterieren. Eine Mglichkeit ist die Verwendung einer for-Schleife. Eine Schleife wiederholt eine bestimmte Aktion mehrere Male. Wir knnen beispielsweise eine Schleife verwenden, um jedes Element eines Arrays zu durchlaufen und das Element auszugeben.

Hier ist ein Beispiel, wie man eine for-Schleife verwenden kann, um die Massen der Planeten in unserem Sonnensystem auszugeben:

````python
# Array mit den Massen der Planeten in Kilogramm
massen = np.array([3.30e23, 4.87e24, 5.97e24, 6.42e23, 1.90e27, 5.68e26, 8.68e25, 1.02e26])

# Iteriere durch das Array und gib jede Masse aus
for masse in massen:
    print("Masse: ", masse)
````

Eine andere Mglichkeit ist die Verwendung von nditer(), um durch ein Array zu iterieren. Hier ist ein Beispiel, wie man nditer() verwenden kann, um die Abstnde der Planeten in unserem Sonnensystem auszugeben:

````python
# Array mit den Abstnden der Planeten in Kilometern
distanzen = np.array([57910000, 108200000, 149600000, 227940000, 778330000, 1429400000, 2870990000, 4497100000])

# Erzeuge einen Iterator, der durch das Array iteriert
iterator = np.nditer(distanzen)

# Iteriere durch das Array und gib jeden Abstand aus
while not iterator.finished:
    print("Abstand: ", iterator[0])
    iterator.iternext()
````

In den folgenden Beispiel verwenden wir nditer() zum Durchlaufen eines zweidimensionalen Arrays mit Temperaturen. Wir geben die Temperatur jedes Elements aus, indem wir den Iterator it verwenden und jedes Mal, wenn wir durchlaufen, it[0] ausgeben. iternext() wird verwendet, um zum nchsten Element im Iterator zu gehen.

````python
# Array mit den Temperaturen in Grad Celsius
temperaturen = np.array([[15, 20, 18],
                         [10, 12, 8],
                         [22, 25, 20]])

# Erzeuge einen Iterator, der durch das Array iteriert
it = np.nditer(temperaturen)

# Iteriere durch das Array und gib die Temperatur jedes Elements aus
while not it.finished:
    print("Temperatur: ", it[0])
    it.iternext()
````

## Numpy Array Suche 

Die NumPy-Array-Suche ermglicht es uns, in einem Array nach bestimmten Elementen zu suchen. Dazu knnen wir verschiedene Funktionen verwenden, wie z.B. np.where(), np.argwhere() oder np.searchsorted(). Diese Funktionen geben uns entweder den Index oder das Element zurck, das der Suche entspricht.

In diesem Beispiel haben wir ein Array distances mit den Entfernungen der Planeten zum Sonnensystem. Mit np.where() suchen wir nach den Planeten, die weiter als 1 Milliarde Kilometer entfernt sind und speichern das Ergebnis in der Variablen entfernte_planets. np.where() gibt uns ein Tupel zurck, das den Index jedes Elements enthlt, das der Suche entspricht.

````python
# Array mit den Entfernungen der Planeten zum Sonnensystem in Millionen Kilometern
distanzen = np.array([57.91, 108.2, 149.6, 227.9, 778.5, 1433.5, 2872.5, 4495.1])

# Suche nach den Planeten, die weiter als 1 Milliarde Kilometer entfernt sind
entfernte_planets = np.where(distanzen > 1000)
print(entfernte_planets)
````

In diesem Beispiel haben wir ein Array diameters mit den Durchmessern der Planeten. Mit np.argwhere() suchen wir nach den Planeten, die grer als 10.000 Kilometer sind und speichern das Ergebnis in der Variablen grosse_planeten. np.argwhere() gibt uns ein Tupel zurck, das den Index jedes Elements enthlt, das der Suche entspricht.

````python
# Array mit den Durchmessern der Planeten in Kilometern
diameter = np.array([4879, 12104, 12756, 6792, 142984, 120536, 51118, 49528])

# Suche nach den Planeten, die grer als 10.000 Kilometer sind
grosse_planets = np.argwhere(diameter > 10000)
print(grosse_planets)
````

In diesem Beispiel haben wir ein Array temperatures mit verschiedenen Temperaturen. Mit np.searchsorted() suchen wir nach der Position der Temperatur 25 Grad Celsius und speichern das Ergebnis in der Variablen pos. np.searchsorted() gibt uns den Index zurck, an dem das Element in das Array eingefgt werden knnte, ohne die Sortierung des Arrays zu verndern.

````python
# Array mit den Temperaturen in Grad Celsius
temperaturen = np.array([10, 15, 20, 25, 30, 35, 40])

# Suche nach der Position der Temperatur 25 Grad Celsius
position = np.searchsorted(temperaturen, 25)
print(position)
````

# Numpy Universal Functions


Numpy Universal Functions (ufuncs) sind Funktionen, die auf jeder Komponente eines Numpy-Arrays angewendet werden knnen. Die ufuncs bieten eine effiziente Mglichkeit, mathematische Operationen auf groen Datenmengen durchzufhren, ohne eine Schleife zu schreiben.

Zum Beispiel knnen ufuncs verwendet werden, um trigonometrische Funktionen wie Sinus, Cosinus und Tangens auf alle Elemente eines Arrays anzuwenden. Ufuncs knnen auch zur Durchfhrung von Arithmetik, zum Finden von Minimum- und Maximumwerten und zur Anwendung von Logarithmen und Exponenten verwendet werden.



## Simple Arithmetik 

Hier sind einige Beispiele fr arithmetische Operationen, die mit ufuncs durchgefhrt werden knnen:

In jedem Beispiel wird die entsprechende ufunc auf die Arrays arr1 und arr2 angewendet und das Ergebnis in der Variable result gespeichert. Die Ausgabe zeigt das Ergebnis der Berechnung fr jedes Element der Arrays.

````python
# Erstelle zwei Arrays
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([5, 6, 7, 8])

# Addition
result = np.add(arr1, arr2)
print(result)  

# Subtraktion
result = np.subtract(arr1, arr2)
print(result)  

# Multiplikation
result = np.multiply(arr1, arr2)
print(result)  

# Division
result = np.divide(arr1, arr2)
print(result)  

# Potenzierung
result = np.power(arr1, 2)
print(result)  

# Quadratwurzel
result = np.sqrt(arr1)
print(result)  

# Exponentialfunktion
result = np.exp(arr1)
print(result) 
````

## Trigonometrie - TBD

Numpy bietet eine Vielzahl von Trigonometrischen ufuncs an, darunter Sinus, Kosinus, Tangens, Arcusfunktionen und mehr. Diese Funktionen knnen in der Astronomie und Raumfahrt hufig verwendet werden, um Berechnungen durchzufhren, wie z.B. die Bestimmung der Position von Himmelskrpern.

**Beispiel 1**: Berechnung des Winkels zwischen zwei Sternen

Angenommen, wir haben die Koordinaten (x, y, z) von zwei Sternen in einem dreidimensionalen Koordinatensystem gegeben. Wir knnen den Winkel zwischen den beiden Sternen mit Hilfe des Skalarprodukts berechnen:

````python
# Koordinaten von Stern 1
x1, y1, z1 = 1, 2, 3

# Koordinaten von Stern 2
x2, y2, z2 = 4, 5, 6

# Skalarprodukt berechnen
dot_produkt = x1*x2 + y1*y2 + z1*z2

# Betrag der Vektoren berechnen
magnitude1 = np.sqrt(x1**2 + y1**2 + z1**2)
magnitude2 = np.sqrt(x2**2 + y2**2 + z2**2)

# Winkel berechnen
winkel = np.arccos(dot_produkt / (magnitude1 * magnitude2))

# Ergebnis ausgeben
print("Der Winkel zwischen den beiden Sternen betrgt:", winkel)
````

**Beispiel 2**: Berechnung der Entfernung zwischen einem Planeten und der Sonne

Angenommen, wir haben die Koordinaten (x, y, z) eines Planeten in einem dreidimensionalen Koordinatensystem gegeben. Wir knnen die Entfernung des Planeten von der Sonne mit Hilfe des Pythagoras-Satzes berechnen:

````python
# Koordinaten des Planeten
x, y, z = 1, 2, 3

# Entfernung zur Sonne berechnen
distanz = np.sqrt(x**2 + y**2 + z**2)

# Ergebnis ausgeben
print("Die Entfernung des Planeten von der Sonne betrgt:", distanz)
````
